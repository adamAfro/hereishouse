# Parametry

```python
import params

df = params.from_file("dane/train.csv")
```

```python
df.to_csv("dane/train.parametry.csv", index=False)
```

```python
# zamiana metrażu na liczby do analizy
from pandas import to_numeric

df["m"] = to_numeric(df["m"])
```

```python
for k in df.columns:
    print(k)
```

## Pieniądze

```python
from matplotlib.pyplot import subplots

fig, axes = subplots(figsize=(8, 8), nrows=2, ncols=2)

axes[0,0].set_title("Rynek")
df["market"].value_counts().plot.pie(ax=axes[0,0], ylabel = "");

axes[1,0].set_title("Waluta zakupu")
df["price[currency]"].value_counts().plot.pie(ax=axes[1,0], ylabel = "");

axes[1,1].set_title("Waluta wynajmu")
df["rent[currency]"].value_counts().plot.bar(ax=axes[1,1], xlabel = "");

axes[0,1].set_title("Waluta depozytu")
df["deposit[currency]"].value_counts().plot.pie(ax=axes[0,1], ylabel = "");
```

## Budowla

```python
from matplotlib.pyplot import subplots

fig, axes = subplots(figsize=(8, 5), nrows=1, ncols=2)

axes[0].set_title("Budowla")
df["building_type"].value_counts().plot.bar(ax=axes[0], xlabel = "");

axes[1].set_title("Typ")
df["type"].value_counts().plot.pie(ax=axes[1], ylabel = "");
```

```python
fig, axes = subplots(figsize=(10, 10), nrows=2, ncols=2)

axes[0,0].set_title("metraż (m)")
df["m"].hist(ax = axes[0,0])

axes[0,1].set_title("m < 500")
df[df["m"] < 500]["m"].hist(ax = axes[0,1]);

axes[1,0].set_title("m < 150")
df[df["m"] < 150]["m"].hist(ax = axes[1,0]);

axes[1,1].set_title("m < 100")
df[df["m"] < 100]["m"].hist(ax = axes[1,1]);
```

```python
ax = to_numeric(df["rooms_num"].replace("more", "-1")).hist();
ax.set_title("Liczba pokoi (-1 znaczy więcej)");
```
