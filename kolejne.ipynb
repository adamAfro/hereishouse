```python
import pandas as pd
import re
dane = pd.read_csv('train.csv')
```

```python
miasto = pd.read_csv('cities.csv')
miasto.rename(columns= {"id" : 'city_id'}, inplace=True)

dzielnica = pd.read_csv('districts.csv')
dzielnica.rename(columns={'id':'district_id'}, inplace=True)

brak_dzielnicy = {'district_id':0, 'name_pl':'', 'lon':0, 'lat':0}
dzielnica.loc[len(dzielnica)] = brak_dzielnicy
```

```python
dane.district_id = dane.district_id.fillna('0')
dane.district_id = dane.district_id.astype(int)
```

```python
dane = pd.merge(dane, miasto, on = 'city_id', how='left')
dane = pd.merge(dane, dzielnica, on = 'district_id', how='left')
```

```python
dane['CityDistrict'] = dane['name_pl_x'] + dane['name_pl_y']
dane
```

```python
dane.drop(columns=['name_pl_y', 'name_pl_x'], inplace=True)
```
