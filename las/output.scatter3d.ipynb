```python
from pandas import DataFrame
```

```python
def read_scores() -> DataFrame:

    from pandas import read_csv, concat

    df = concat([
        
        read_csv('output@kaggle/GridSearchCV1%.csv'),
        read_csv('output@kaggle/GridSearchCV5%.csv'),
        read_csv('output@kaggle/GridSearchCV5%.2.csv'),
        read_csv('output@kaggle/GridSearchCV5%.3.csv')

    ],  ignore_index=True)
    
    df.columns = [col.replace('param_', '') if col.startswith('param_') else col for col in df.columns]

    return df

df = read_scores()
df.sort_values(by='mean_test_score', ascending=True)
df
```

```python
from scatter3d import scatter_3d_params
scatter_3d_params(df[df['max_samples'] == 0.05], color_scale=[0,0.25])
```
