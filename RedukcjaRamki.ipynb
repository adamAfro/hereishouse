## Pozbywanie się nieostotnych zmiennych

### Naprawa poprzedniej ramki danych
 - dodanie kolumn created_at_first_year, created_at_first_month, created_at_first_day
 - zamiana typu zmiennej created_at_first na datetime
 - zamiana niektórych zmiennych na binarne
 - zamiana zmiennych typu object na string

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:\\Users\\48509\\Desktop\\V_semestr\\ATLAS\\train.no-text.csv', low_memory = False)
```

```python
pd.set_option('display.max_columns', 120)
df.head()
```

```python
# dodanie kolumn rok, miesiac, dzien i zamiana typu kolumny created_at_first

nowa = pd.to_datetime(df["created_at_first"])
df["created_at_first_year"] = nowa.dt.year
df["created_at_first_month"] = nowa.dt.month
df["created_at_first"] = nowa
```

```python
# Znajdź kolumny, które mają tylko jedną unikalną wartość
single_value_columns = df.columns[df.nunique() == 1]

# Zamień te kolumny na binarne, gdzie NaN jest False
for col in single_value_columns:
    df[col] = df[col].notna()
```

## Zredukowanie ilości kolumn w ramce danych

```python
# kolumny z wartościami brakującymi posortowane malejąco

columns_with_missing_values = df.columns[df.isna().any()]
sorted_columns = df[columns_with_missing_values].isna().sum().sort_values(ascending=False).index
result_df = df[sorted_columns]

print(result_df)
```
