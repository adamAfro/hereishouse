## Ramka z długościami tekstów i danymi 01

```python
from pandas import read_csv

df = read_csv("dane/train.desc-len.csv")
```

## Data

```python
import pandas as pd

nowa = pd.to_datetime(df["created_at_first"])
nowa.head()

df["created_at_first"] = nowa
```

## Miasto i dzielnica

```python
miasto = pd.read_csv('dane/cities.csv')
miasto.rename(columns= {"id" : 'city_id'}, inplace=True)

dzielnica = pd.read_csv('dane/districts.csv')
dzielnica.rename(columns={'id':'district_id'}, inplace=True)

brak_dzielnicy = {'district_id':0, 'name_pl':'', 'lon':0, 'lat':0}
dzielnica.loc[len(dzielnica)] = brak_dzielnicy

df.district_id = df.district_id.fillna('0')
df.district_id = df.district_id.astype(int)

df = pd.merge(df, miasto, on = 'city_id', how='left')
df = pd.merge(df, dzielnica, on = 'district_id', how='left')

df['CityDistrict'] = df['name_pl_x'] + df['name_pl_y']

#df.drop(columns=['name_pl_y', 'name_pl_x'], inplace=True)
```

```python
df.drop(columns=['name_pl_y'], inplace=True) # zostawiam miasto
df.rename(columns={'name_pl_x':'city'}, inplace=True)
df.rename(columns={'CityDistrict':'city_district'}, inplace=True)
```

```python
df.rename(columns={
    'lon_x':'lon_city',
    'lat_x':'lat_city',
    'lon_y':'lon_district',
    'lat_y':'lat_district'
}, inplace=True)
```

```python
df.drop(columns=['district_id', 'city_id', 'region_id'], inplace=True)
```

## Parametry

```python
df_par = read_csv("dane/train.par.csv")
```

```python
from pandas import concat

df = concat([df, df_par], axis=1)
df.drop(columns=['params'], inplace=True)
```

## Ramka końcowa

```python
df.drop(columns=['id'], inplace=True)
df
```

```python
df.to_csv('dane/train.no-text.csv', index=False)
```
